```python
import numpy as np
import cmath
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft

# Frequences definition

f0 = 10  # Hz
f1 = f0
f2 = 2 * f0
f3 = f0 / 2
fs = 600*(f2-f3)  # Hz (czstotliwo prbkowania)
T = 20 / (f1-f3)  # cakowity czas
t = np.arange(0, T, 1/fs) # time vector

# Signal function of time
signal_t = 2*np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t) + 3*np.sin(2 * np.pi * f3 * t)

plt.plot(t,signal_t)
```

```python
# DFT:

N = len(t)
k = np.arange(0, 8.5*f2, 1/(f0)) * fs / N


# Signal function of frequences
signal_freq = 2 * np.abs([complex(np.sum([signal_t[i] * np.exp(-1j*2*np.pi*kj*i/N) for i in range(N)])) for kj in k]) / N
```

```python
plt.plot(k[0:len(k)//2], signal_freq[0:len(signal_freq)//2])
```

```python
fs = 50
T = 1
N = T * fs
t = np.arange(0, N, 1) / fs

signal_t =  2*np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t) + 3*np.sin(2 * np.pi * f3 * t)

plt.plot(t, signal_t)
plt.scatter(t, signal_t)
plt.show
```

```python
# FFT

signal_freq = fft(signal_t)

freq = fftfreq(N, 1 / fs)[:N//2]

amplitudes = 2.0/N * np.abs(signal_freq[:N//2])

plt.plot(freq, amplitudes)
plt.scatter(freq, amplitudes)
plt.show
```
