```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar
from scipy.special import hermite, factorial

# Parametry studni
x0 = 2.5                # półszerokość studni
#V_prog = 1/2 * x**2           # oscylator harmoniczny
n = 60                   # numer stanu

E = n + 0.5

def psi(x, n=n):
    Hn = hermite(n)              # wielomian Hermite’a stopnia n
    norm = 1.0 / (np.pi**0.25 * np.sqrt(2**n * factorial(n)))
    return norm * np.exp(-x**2 / 2) * Hn(x)

# Siatka x
x = np.linspace(-5*x0, 5*x0, 1000)
```

```python
# Rysuj funkcję falową
psi_vals = psi(x)

plt.figure(figsize=(10, 5))
plt.plot(x, np.abs(psi_vals)**2, label=f"$|\psi_{n}|^2$")

plt.axvline(-x0, color='gray', linestyle='--')
plt.axvline(x0, color='gray', linestyle='--')
plt.title(f"Funkcja falowa dla $E$ = {E:.1f}, $n$ = {n}")
plt.xlabel("$x$")
plt.ylabel("$|ψ(x)|^2$")
plt.legend()
plt.grid(True)
plt.tight_layout()
#plt.xlim(-2*x0, 2*x0)
plt.show()
```
